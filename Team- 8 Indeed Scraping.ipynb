```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import re, time,csv
import csv


# Opens Chrome driver in full screen and applies the setting , disable javascript and runs on incognito
driver = webdriver.Chrome('./chromedriver')
chrome_options = Options()
chrome_options.add_argument("--disable-javascript")
chrome_options.add_argument("--incognito")
driver.maximize_window()

# Remove non-alphabets including punctuations and numbers from job description
pattern = re.compile(r'[^a-zA-Z]', re.UNICODE)


with open("indeed_jobs.csv", "a", encoding='utf-8', newline='') as file:
#with open("sw6.csv", "a", encoding='utf-8', newline='') as file:
    csv_writer = csv.writer(file, delimiter=',')
    csv_writer.writerow(["text", "title"])
    #for job_type in ["software+engineer"]:
    #for job_type in ["Data+Scientists"]:
    for job_type in ["Data+Scientist", "software+engineer"]:
        
        
        # Number of pages, here we used 58 (each page has 15 jobs listed)
        for page in range(0, 58): 
            page_url = driver.current_url
            #page_url = "https://www.indeed.com/jobs?q="+ job_type +"&l=United+States&start=" + str(page*10)
            page_url = "https://www.indeed.com/jobs?q="+ job_type +"&l=Texas&start=" + str(page*10)
            driver.get(page_url)
        
            time.sleep(2)
            ### Closes and re-opens the web browser driver and goes to the link
            if (page+1)%5 ==0:
                    driver.close()
                    driver = webdriver.Chrome('./chromedriver')
                    chrome_options = Options()
                    chrome_options.add_argument("--disable-javascript")
                    chrome_options.add_argument("--incognito")
                    driver.maximize_window()
                    driver.get(page_url)
                    print("Wait time : 10s")
                    time.sleep(30)
            
            
            rows = driver.find_elements_by_css_selector('[class="slider_container css-11g4k3a eu4oa1w0"]')
            

            for row in rows:
                driver.switch_to.default_content()
                time.sleep(1)
                ### Click every job box
                try:
                    row.click()
                    time.sleep(2)
                except:
                    print("job box error")
                    continue
                
                ### Closes pop_up
                try:
                    close_popup = driver.find_element_by_id("popover-x")
                    close_popup.click()
                    time.sleep(1)
                except:
                    try:
                        close_popup = driver.find_elements_by_css_selector('[class="popover-x-button-close icl-CloseButton"]')
                        close_popup.click()
                        time.sleep(1)
                    except:
                        print("No Popup Found !")
                

                ### Closes pop_up
                try:
                    close_popup2 = driver.find_elements_by_css_selector('[ class="icl-CloseButton icl-Modal-close"]')
                    close_popup2.click()
                except:
                    try:
                        close_popup2 = driver.find_elements_by_class_name("icl-CloseButton icl-Modal-close")
                        close_popup2.click()
                    except:
                        print()
                ## Get Job Description
                try:
                    driver.switch_to.frame(driver.find_element_by_id("vjs-container-iframe"))
                    content = driver.find_element_by_id('jobDescriptionText')
                    job_desc = content.text
                    job_desc = content.text
                    # remove non-alphabets
                    job_desc = pattern.sub(' ', job_desc)
                    job_desc = re.sub(r'\s+', ' ', job_desc).strip()
                    # Convert to lowercase
                    job_desc = job_desc.lower()

                    # write to csv
                    csv_writer.writerow([job_desc, job_type.replace("+", " ")])
                except:
                    print('description error')
            
                
            ## Print Page
            print('Page number ', page+1, ' read')
            

driver.close()
```
